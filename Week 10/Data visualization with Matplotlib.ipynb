```python
import matplotlib.pyplot as plt 
```

```python
day = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
steps_walked = [1700, 2000, 5000, 3500, 9000, 7000, 8000]
steps_walked_last_week = [9000, 23000, 10000, 4000, 6000, 8000, 7500]

plt.plot(day, steps_walked, "x--m")
plt.plot(day, steps_walked_last_week, "o-g")
plt.title("Steps walked for each day")
plt.xlabel("Days of the week")
plt.ylabel("Step count")
plt.legend(["This week", "Last week"])
plt.grid(True)
plt.show()
```
